```python
# QAI_Agile_Product_Owner Process Simulation (Colab Ready)

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# ============================
# 1. Sample Client Order
# ============================
client_order = {
    "Client": "QuantumTech Industries",
    "Order": "Hybrid QAI Processor + OS Prototype",
    "Budget": "2M USD",
    "Deadline": "6 months",
    "Strategic Goals": ["Post-Quantum Security", "High Fidelity > 99%", "Low Cryogenic Power Usage"]
}

print("=== Sample Client Order ===")
for k, v in client_order.items():
    print(f"{k}: {v}")

# ============================
# 2. Define Lifecycle Steps, Milestones
# ============================
lifecycle = [
    "Business & Strategy",
    "Planning & Roadmap",
    "Agile Execution",
    "Delivery & Feedback",
    "Systems & Blueprint",
    "Manufacturing & Prototype",
    "Continuous Learning"
]

milestones = ["Vision Charter", "PI Planning", "First Increment", "Prototype", "Major Release"]

# ============================
# 3. Dummy KPI Data
# ============================
kpis = {
    "Business": 80,
    "Planning": 70,
    "Execution": 85,
    "Delivery": 75,
    "Blueprint": 65,
    "Manufacturing": 60,
    "Learning": 90
}

# ============================
# 4. Dummy Manufacturing Yield Data
# ============================
cycles = np.arange(1,6)
yield_rate = [50, 65, 70, 80, 85]

# ============================
# 5. Swimlane Chart
# ============================
iterations = ["Cycle 1", "Cycle 2", "Cycle 3", "Cycle 4"]

fig, ax = plt.subplots(figsize=(10,6))
for i, layer in enumerate(lifecycle):
    ax.plot(iterations, [i+1]*len(iterations), marker="o", label=layer)
ax.set_yticks(range(1, len(lifecycle)+1))
ax.set_yticklabels(lifecycle)
ax.set_title("Swimlane: Layers vs Iteration Cycles")
ax.grid(True, axis="x", linestyle="--", alpha=0.7)
plt.legend(bbox_to_anchor=(1.05,1), loc="upper left")
plt.show()

# ============================
# 6. KPI Dashboard
# ============================
fig, ax = plt.subplots(figsize=(8,5))
ax.bar(kpis.keys(), kpis.values(), color="#66b2ff")
ax.set_ylabel("KPI Achievement (%)")
ax.set_title("KPI Dashboard (Sample Data)")
plt.xticks(rotation=30)
plt.show()

# ============================
# 7. Manufacturing Yield Trend
# ============================
fig, ax = plt.subplots(figsize=(8,5))
ax.plot(cycles, yield_rate, marker="o", linestyle="-", color="green")
ax.set_xticks(cycles)
ax.set_xlabel("Prototype Iteration")
ax.set_ylabel("Yield (%)")
ax.set_title("Manufacturing Yield Across Iterations")
plt.show()

# ============================
# 8. Release Milestones Timeline
# ============================
timeline = [1, 2, 3, 4, 5]
fig, ax = plt.subplots(figsize=(10,2))
ax.hlines(1, 1, 5, color="gray", linewidth=2)
ax.plot(timeline, [1]*len(timeline), "o", color="blue")
for x, label in zip(timeline, milestones):
    ax.text(x, 1.05, label, ha="center", fontsize=9, weight="bold")
ax.set_ylim(0.8, 1.2)
ax.axis("off")
ax.set_title("Release & Milestones Timeline")
plt.show()

print("\n=== Process Simulation Complete ===")
print("Generated visuals for Swimlane, KPI Dashboard, Yield Trend, and Milestone Timeline.")
```
