```python
# QAI Client Adoption Tool ‚Äî Unified End-to-End Demo (Colab)
# Combines: charter ‚Üí placeholders ‚Üí product bundle ‚Üí roadmap ‚Üí ROI
# with program/projects, governance/RACI, agile burnup + CRs, EV/PV/AC,
# TRL/CMMI maturity, current‚Üífuture mapping, dependency DAG, approvals, KPIs.

# ‚îÄ‚îÄ Setup ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
import sys, math, itertools, datetime as dt
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# Ensure networkx for dependency graph (lightweight)
try:
    import networkx as nx
except ImportError:
    !pip -q install networkx
    import networkx as nx

pd.set_option("display.max_colwidth", 200)

# ‚îÄ‚îÄ SECTION A: Charter, Placeholders, QAI Bundle, Roadmap, ROI ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
client_charter = {
    "Project": "Hybrid QAI Datacenter Modernization",
    "Domain": "Enterprise IT",
    "Budget_USD": 2_000_000,
    "Scope": ["Datacenter Upgrade", "Security Compliance", "Hybrid Cloud"],
    "Constraints": {"Data Residency": None, "SLA_ms": None},
}
print("üìÑ Client Project Charter")
for k, v in client_charter.items():
    print(f" - {k}: {v}")

placeholders = pd.DataFrame([
    {"Field": "Data Residency", "Owner": "Compliance Officer", "Status": "Open"},
    {"Field": "SLA (ms)", "Owner": "Client IT Lead", "Status": "Open"},
])
print("\n‚ö†Ô∏è Placeholders")
display(placeholders)

qai_bundle = pd.DataFrame([
    {"QAI Product": "QAI Datacenter", "Purpose": "Software-defined hybrid datacenter",
     "Dependencies": "10/40G fabric, SDN", "Status": "Compatible"},
    {"QAI Product": "QAI OS", "Purpose": "Hybrid quantum-classical runtime",
     "Dependencies": "Linux + Container runtime", "Status": "Compatible"},
    {"QAI Product": "QAI Security Suite", "Purpose": "Post-Quantum Security + IAM",
     "Dependencies": "HSM, IAM provider", "Status": "Upgrade Needed"},
])
print("\nüß© Recommended QAI Product Bundle")
display(qai_bundle)

roadmap = pd.DataFrame([
    {"Phase": "Intake",                    "TRL": 3, "CMMI": 1, "Output": "Scope Baseline, Placeholders"},
    {"Phase": "Compliance & Infra Scan",   "TRL": 4, "CMMI": 2, "Output": "Gap Report, Risk Register"},
    {"Phase": "Product Match",             "TRL": 5, "CMMI": 2, "Output": "Bundle & Dependency Map"},
    {"Phase": "Pilot",                     "TRL": 6, "CMMI": 3, "Output": "Prototype, Test Results"},
    {"Phase": "Deploy",                    "TRL": 7, "CMMI": 3, "Output": "Live Environment, SLO Monitors"},
    {"Phase": "Operate",                   "TRL": 8, "CMMI": 4, "Output": "Ops Handover Pack, Continuous Improvement"},
])
print("\nüöÄ Adoption Roadmap (with maturity)")
display(roadmap)

roi_baseline = {
    "Current OPEX/year": 1_200_000,
    "Projected OPEX/year (QAI)": 850_000,
}
roi_baseline["Annual Savings"] = roi_baseline["Current OPEX/year"] - roi_baseline["Projected OPEX/year (QAI)"]
roi_baseline["Payback Period (years)"] = round(client_charter["Budget_USD"] / max(1, roi_baseline["Annual Savings"]), 2)
print("\nüí∞ ROI Estimate")
for k, v in roi_baseline.items():
    print(f" - {k}: {v}")

# ‚îÄ‚îÄ SECTION B: Program, Projects, Governance, RACI ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
program = {
    "Program": "Hybrid QAI Datacenter Modernization",
    "Sponsor": "Client CIO",
    "Change Manager": "Startup Change Lead",
    "Duration_weeks": 32,
    "Objective": "Modernize DC with QAI stack, ensure PQC compliance, improve OPEX"
}
print("\nüìå Program Summary")
for k, v in program.items():
    print(f" - {k}: {v}")

projects = pd.DataFrame([
    {"ProjectID":"P-001", "Name":"Core DC Upgrade",              "Type":"Brownfield", "Status":"In Progress", "Budget":1_400_000},
    {"ProjectID":"P-002", "Name":"Security & PQC Enablement",    "Type":"Greenfield", "Status":"In Progress", "Budget":350_000},
    {"ProjectID":"P-003", "Name":"Robotics & OT Integration",    "Type":"Pilot",      "Status":"Planned",     "Budget":250_000},
])
display(projects)

phases = pd.DataFrame([
    {"Phase":"Intake",              "Start_wk":0,  "Duration_wk":2,  "Gate":"Scope Baseline",         "Gov":"Plan",  "TRL_before":2, "TRL_after":3, "CMMI_before":1, "CMMI_after":1},
    {"Phase":"Infra Discovery",     "Start_wk":2,  "Duration_wk":3,  "Gate":"Compat Report Approved", "Gov":"Do",    "TRL_before":3, "TRL_after":4, "CMMI_before":1, "CMMI_after":2},
    {"Phase":"Compliance Mapping",  "Start_wk":5,  "Duration_wk":3,  "Gate":"Controls & Exceptions",  "Gov":"Check", "TRL_before":4, "TRL_after":4, "CMMI_before":2, "CMMI_after":2},
    {"Phase":"Product Match",       "Start_wk":8,  "Duration_wk":3,  "Gate":"Design Approval",        "Gov":"Act",   "TRL_before":4, "TRL_after":5, "CMMI_before":2, "CMMI_after":2},
    {"Phase":"Roadmap & Pilot",     "Start_wk":11, "Duration_wk":8,  "Gate":"Pilot Exit/Go",          "Gov":"Do",    "TRL_before":5, "TRL_after":6, "CMMI_before":2, "CMMI_after":3},
    {"Phase":"Deploy",              "Start_wk":19, "Duration_wk":8,  "Gate":"Go-Live",                "Gov":"Act",   "TRL_before":6, "TRL_after":7, "CMMI_before":3, "CMMI_after":3},
    {"Phase":"Operate & Improve",   "Start_wk":27, "Duration_wk":5,  "Gate":"Handover & SLOs Met",    "Gov":"Check", "TRL_before":7, "TRL_after":8, "CMMI_before":3, "CMMI_after":4},
])
phases["End_wk"] = phases["Start_wk"] + phases["Duration_wk"]

print("\nüß≠ Lifecycle Phases & Gates")
display(phases[["Phase","Start_wk","End_wk","Gate","Gov","TRL_before","TRL_after","CMMI_before","CMMI_after"]])

raci = pd.DataFrame([
    {"Activity":"Scope Baseline",                 "R":"Change Manager",   "A":"Client Sponsor", "C":"Solution Architect", "I":"Finance, Legal"},
    {"Activity":"Infra Compat Report",            "R":"Infra Agent",      "A":"Client IT Lead", "C":"Security Lead",      "I":"Vendors"},
    {"Activity":"Compliance Exceptions",          "R":"Compliance Agent", "A":"Compliance Officer", "C":"Legal",        "I":"Auditors"},
    {"Activity":"QAI Bundle Selection",           "R":"QAI Matchmaker",   "A":"Solution Architect", "C":"Product Owners","I":"Partners"},
    {"Activity":"Pilot Exit",                     "R":"PMO",              "A":"Client Sponsor", "C":"Ops Manager",        "I":"End Users"},
    {"Activity":"Go-Live",                        "R":"Ops Manager",      "A":"Client CIO",     "C":"Security Lead",      "I":"All"},
    {"Activity":"Ops Handover & SLO Acceptance",  "R":"Ops Agent",        "A":"Client Ops Head","C":"PMO",                "I":"Auditors"},
])
print("\nüßë‚Äçü§ù‚Äçüßë Governance RACI")
display(raci)

# ‚îÄ‚îÄ SECTION C: Agile Burnup + Change Requests ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
weeks = np.arange(0, program["Duration_weeks"]+1)
sprint_len = 2
total_scope_sp = 420
cr_events = {6:+30, 13:+20, 18:-15, 22:+25}  # week: delta scope
scope_series = []
current_scope = total_scope_sp
for w in weeks:
    current_scope += cr_events.get(w, 0)
    scope_series.append(current_scope)

np.random.seed(42)
velocity_per_week = 12
noise = np.random.normal(0, 6, size=len(weeks))
done_raw = (weeks * velocity_per_week) + noise
done_raw = np.clip(done_raw, 0, None)
completed = np.minimum(scope_series, np.maximum.accumulate(done_raw))
burnup_df = pd.DataFrame({"Week":weeks, "Scope_SP":scope_series, "Done_SP":completed})

# ‚îÄ‚îÄ SECTION D: Budget Tracking (EV/PV/AC) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
months = pd.date_range("2025-01-01", periods=8, freq="MS")
baseline_total = projects["Budget"].sum()  # 2.0M
pv = np.linspace(0, baseline_total, len(months))
week_per_month = program["Duration_weeks"]/len(months)
ev = []
ac = []
for i in range(len(months)):
    wk = int(min(len(weeks)-1, round((i+1)*week_per_month)))
    ratio = burnup_df.loc[wk, "Done_SP"] / max(1, burnup_df.loc[wk, "Scope_SP"])
    ev.append(baseline_total * ratio)
    ac.append(max(0, pv[i] + np.random.normal(0, baseline_total*0.03)))
ev = np.array(ev); ac = np.array(ac)
ev_df = pd.DataFrame({"Month":months, "PV":pv, "EV":ev, "AC":ac})
ev_df["CPI"] = np.where(ev_df["AC"]>0, ev_df["EV"]/ev_df["AC"], np.nan)
ev_df["SPI"] = np.where(ev_df["PV"]>0, ev_df["EV"]/ev_df["PV"], np.nan)
print("\nüí∞ EV / PV / AC (Monthly)")
display(ev_df)

# ‚îÄ‚îÄ SECTION E: Current ‚Üí Future Mapping ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
mapping = pd.DataFrame([
    {"Dimension":"ERP",        "Current":"SAP ECC on-prem",               "Future":"Hybrid ERP + QAI OS connectors",           "Compliance":"ISO 27001", "Value":"Faster integrations"},
    {"Dimension":"Security",   "Current":"RSA-2048, mixed IAM",           "Future":"PQC suite + unified IAM",                  "Compliance":"NIST PQC",  "Value":"Long-term cryptographic safety"},
    {"Dimension":"Datacenter", "Current":"Legacy servers, high PUE",      "Future":"Software-defined QAI Datacenter",          "Compliance":"ISO 50001", "Value":"Lower OPEX"},
    {"Dimension":"Robotics",   "Current":"Minimal automation",            "Future":"QAI Robots (semi-autonomous)",             "Compliance":"ISO 13482", "Value":"Throughput, safety"},
    {"Dimension":"Ops",        "Current":"Manual runbooks, low SLOs",     "Future":"QAI Ops with SLO mgmt & DR drills",        "Compliance":"ISO 20000", "Value":"Reliability"},
])
print("\nüß© Current ‚Üí Future Mapping")
display(mapping)

# ‚îÄ‚îÄ SECTION F: Dependency Graph (DAG) ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
deps_edges = [
    ("Intake","Infra Discovery"),
    ("Infra Discovery","Compliance Mapping"),
    ("Compliance Mapping","Product Match"),
    ("Product Match","Roadmap & Pilot"),
    ("Roadmap & Pilot","Deploy"),
    ("Deploy","Operate & Improve"),
]
G = nx.DiGraph(); G.add_edges_from(deps_edges)

# ‚îÄ‚îÄ SECTION G: Charts ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
# A) Timeline (Gantt-style)
plt.figure(figsize=(10,4))
for i, row in phases.iterrows():
    plt.barh(row["Phase"], row["Duration_wk"], left=row["Start_wk"], edgecolor="black")
    plt.text(row["Start_wk"] + row["Duration_wk"]/2, i, row["Gate"], ha="center", va="center", fontsize=8)
plt.xlabel("Week")
plt.title("Lifecycle Timeline (Phases & Gates)")
plt.tight_layout()
plt.show()

# B) Burnup with CR markers
plt.figure(figsize=(10,4))
plt.plot(burnup_df["Week"], burnup_df["Scope_SP"], label="Total Scope (SP)")
plt.plot(burnup_df["Week"], burnup_df["Done_SP"], label="Done (SP)")
for w, d in cr_events.items():
    plt.axvline(w, linestyle="--", alpha=0.5)
    plt.text(w+0.2, max(burnup_df["Scope_SP"])*0.05, f"CR {d:+}", rotation=90, fontsize=8)
plt.xlabel("Week"); plt.ylabel("Story Points")
plt.title("Agile Burnup (Scope vs Done) + Change Requests")
plt.legend()
plt.tight_layout()
plt.show()

# C) EV/PV/AC curves
plt.figure(figsize=(10,4))
plt.plot(ev_df["Month"], ev_df["PV"], label="PV")
plt.plot(ev_df["Month"], ev_df["EV"], label="EV")
plt.plot(ev_df["Month"], ev_df["AC"], label="AC")
plt.title("Budget Tracking: PV / EV / AC")
plt.xlabel("Month"); plt.ylabel("USD")
plt.legend()
plt.tight_layout()
plt.show()

print("\nüìà CPI/SPI Snapshot")
display(ev_df[["Month","CPI","SPI"]])

# D) TRL/CMMI progression
plt.figure(figsize=(10,4))
plt.plot(phases["Phase"], phases["TRL_after"], marker="o", label="TRL after")
plt.plot(phases["Phase"], phases["CMMI_after"], marker="o", label="CMMI after")
plt.title("Maturity Progression (TRL & CMMI)")
plt.xlabel("Phase"); plt.ylabel("Level")
plt.xticks(rotation=20)
plt.legend()
plt.tight_layout()
plt.show()

# E) Dependency DAG
plt.figure(figsize=(7,5))
pos = nx.spring_layout(G, seed=7)
nx.draw(G, pos, with_labels=True, node_size=1500, font_size=9, arrows=True)
plt.title("Phase Dependencies (DAG)")
plt.tight_layout()
plt.show()

# ‚îÄ‚îÄ SECTION H: Approvals Queue & KPI Rollup ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ‚îÄ
approvals = pd.DataFrame([
    {"Item":"Scope Baseline","Role":"Change Manager","Status":"Awaiting"},
    {"Item":"Compat Report","Role":"Client IT Lead","Status":"Awaiting"},
    {"Item":"Compliance Exceptions","Role":"Compliance Officer","Status":"Approved"},
    {"Item":"Design Package v1","Role":"Solution Architect","Status":"Awaiting"},
    {"Item":"Pilot Exit","Role":"Client Sponsor","Status":"Planned"},
    {"Item":"Go-Live","Role":"Client CIO","Status":"Planned"},
])
print("\nüìù Approvals Queue")
display(approvals)

value = {
    "Projected OPEX (current ‚Üí QAI)": "1.2M ‚Üí 0.85M (USD/yr)",
    "Availability Target": "99.9%",
    "Latency P95 Target": "‚â§ 120 ms",
    "Energy Efficiency": "‚Üì PUE via QAI DC",
    "Security": "PQC rollout; unified IAM",
}
print("\nüéØ Value & KPI Targets")
for k, v in value.items():
    print(f" - {k}: {v}")

print("\n‚úÖ Unified demo complete: charter ‚Üí bundle ‚Üí roadmap ‚Üí ROI + program/projects, governance,"
      " agile change mgmt, EV/PV/AC, maturity, mapping, dependencies, approvals, and KPIs.")
```
